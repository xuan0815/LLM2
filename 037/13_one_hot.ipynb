#**One-Hot Encoding 獨熱碼**



*   one-hot encoding是用來向量表示字詞的方式，特點是每一個字詞之間皆為正交(orthogonal)
*   字詞向量正交，才能排除字詞之間互相影響
*   one-hot的優點也是缺點，不同詞的向量正交，無法衡量不同詞之間的關係
*   只能反映某個詞是否在句中出現，無法衡量詞的重要性
*   one-hot是高維稀疏矩陣，浪費計算能力與儲存空間
*   本範例示範如何把國家表示成one-hot encoding表示法















```python
# 匯入所需的函式庫
import numpy as np
import pandas as pd

# 創建資料
country = ['台灣', '澳大利亞', '愛爾蘭', '澳大利亞', '愛爾蘭', '台灣']
age = [25, 30, 45, 35, 22, 36]
salary = [20000, 32000, 59000, 60000, 43000, 52000]

# 使用字典創建 DataFrame
dic = {'Country': country, 'Age': age, 'Salary': salary}
data = pd.DataFrame(dic)

# 顯示資料集
print(data)
```

#**Label encoding**


####利用LabelEcoder函數，將Country欄位進行Label

####encoding編碼
####Australia：0
####Ireland：1
####Taiwan：2

```python
# 匯入所需的函式庫
from sklearn.preprocessing import LabelEncoder
# 創建一個 LabelEncoder 物件
labelencoder = LabelEncoder()
# 將資料轉換成 DataFrame
data_le=pd.DataFrame(dic)
# 對 'Country' 欄位進行標籤編碼
data_le['Country'] = labelencoder.fit_transform(data['Country'])
# 顯示標籤編碼後的資料
data_le
```

#**One hot encoding**
######利用OneHotEcoder函數，將Country欄位進行Ont-Hot encoding編碼

######     Australia(0)	Ireland(1)	Taiwan(2)


```python
# 匯入所需的函式庫
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# 創建一個 ColumnTransformer 物件，將第一欄的資料進行獨熱編碼，並保留其餘欄位不變
ct = ColumnTransformer([("Country", OneHotEncoder(), [0])], remainder = 'passthrough')
# 將獨熱編碼應用到資料集 data_le 上
data_str_ohe = ct.fit_transform(data_le)
# 將編碼後的資料轉換成 DataFrame 並顯示
pd.DataFrame(data_str_ohe)
```

#**Pandas.get_dummies**

```python
data_dum = pd.get_dummies(data) # 將資料進行獨熱編碼
pd.DataFrame(data_dum) # 將編碼後的資料轉換成 DataFrame 並顯示
```
